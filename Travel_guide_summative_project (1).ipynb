```python
%%html
<!DOCTYPE html>
<html>
<head>
<title>Paris Travel Guide</title>
</head>

<body>

<h1>Paris Travel Guide</h1>

<p>Paris is the capital of France. It is known as the city of light.
There is a lot of history here and the food is really good.
It is one of the most visited cities in the whole world.</p>

<h2>Things To Do</h2>

<ul>
  <li>Visit the Eiffel Tower - the most famous thing in Paris</li>
  <li>Go to the Louvre - it has the Mona Lisa and tons of other art</li>
  <li>Walk around Montmartre - its a cool artsy neighborhood on a hill</li>
  <li>See Notre Dame Cathedral - its being fixed after the fire in 2019</li>
</ul>

<h2>Food and Culture</h2>

<p>Paris has some of the best food in the world. Here are some things you should try:</p>

<ul>
  <li>Croissants - get one from a real bakery not a tourist trap</li>
  <li>Crepes - you can get these from street carts everywhere</li>
  <li>Baguette with cheese - simple but really good</li>
  <li>Macarons - very sweet but famous</li>
  <li>Steak frites - steak with fries its a classic french meal</li>
</ul>

<p>French people say Bonjour when they walk into stores. You should do it too, they really appreciate it.</p>

<h2>Travel Tips</h2>

<p>Best time to go is Spring (April to June) or Fall (September to October).
Summer is really crowded and hot. August is especially bad because a lot
of Parisian shops and restaurants close while locals go on vacation.</p>

<p>Getting around: The Paris Metro is the easiest way to get around.
It goes pretty much everywhere and is not too expensive.
You can also walk a lot because a lot of the main spots are close together.</p>

<ul>
  <li>Carry some cash because not all places take cards</li>
  <li>Watch your pockets near tourist spots like the Eiffel Tower</li>
  <li>Try to say Bonjour and Merci, locals like it</li>
  <li>Tap water is safe to drink so bring a water bottle</li>
</ul>

</body>
</html>
```
